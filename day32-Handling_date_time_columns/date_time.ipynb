```python
import pandas as pd
import numpy as np
```

```python
date=pd.read_csv('orders.csv')
```

```python
date.info()
```

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#Extracting year,month,days
date['year']=date['date'].dt.year

date
```

```python
date['month']=date['date'].dt.month  # for month 
date['month_name']=date['date'].dt.month_name() # for month name

date
```

```python
date['day']=date['date'].dt.day
date['dayofweek']=date['date'].dt.dayofweek # for the day of the week
date['dayname']=date['date'].dt.day_name() # for the day of the week name
date['weekofyear']=date['date'].dt.isocalendar().week # week of the year
date['quaterofyear']=date['date'].dt.quarter # quater of the year
date.head()
```

```python
# is weekend?

date['is_weekend']=np.where(date['dayname'].isin(['Sunday','Saturday']),1,0)

date
```

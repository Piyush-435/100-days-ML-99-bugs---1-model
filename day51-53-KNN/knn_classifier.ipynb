```python
import numpy as np
import statistics
```

```python
class KNN_classifier():
    
    def __init__(self,distance_metric):
        self.distance_metric=distance_metric
        
        
    def get_distance_metric(self,p1,p2):
        if(self.distance_metric=='eucledian'):
            dist=0
            for i in range(len(p1)-1):
                dist+=(p1[i]-p2[i])**2
            
            eucledian_dist=np.sqrt(dist)
            return eucledian_dist
        elif(self.distance_metric=='manhattan'):
            dist=0
            for i in range(len(p1)-1):
                dist+=abs(p1[i]-p2[i])
            manhattan_dist=dist
            return manhattan_dist
            
            
    
    def nearest_ngh(self,X_train,test_data,k):
        distance_metric=[]
        for training_data in X_train: 
            distance=self.get_distance_metric(training_data,test_data)
            distance_metric.append([training_data,distance])
            
        distance_metric.sort(key=lambda x:x[1])
        
        
        neighbour_list=[]
        for i in range(k):
            neighbour_list.append(distance_metric[i][0])
            
        return neighbour_list
            
    
    
    def predict(self,X_train,test_data,k):
        neighbours=self.nearest_ngh(X_train,test_data,k)
        label=[]
        for data in neighbours:
            label.append(data[-1])
            
        predicted_class=statistics.mode(label)
        return predicted_class
            
            
```

```python
X=df.drop(['Outcome'],axis=1)
y=df['Outcome']

X=X.to_numpy()
y=y.to_numpy()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
X_train=np.insert(X_train,8,y_train,axis=1)
```

x_test->test data without output

```python
classifier=KNN_classifier(distance_metric='eucledian')
```

```python
# Our class can only predict for one point at a time

prediction=classifier.predict(X_train,X_test[0],k=5)
```

```python
# prediction for multiple points

y_pred=[]
for i in range(X_test.shape[0]):
    prediction=classifier.predict(X_train,X_test[i],k=5)
    y_pred.append(prediction)
    
```

```python
accuracy_score(y_test,y_pred)*100
```
